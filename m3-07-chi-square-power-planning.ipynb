### Task 1: Chi-Square Goodness-of-Fit Test 

```python
observed = np.array([120, 85, 110, 85]) 
```

```python
expected = np.array([100, 100, 100, 100]) 
```

```python
chi2_stat, p_value = st.chisquare(observed, f_exp = expected)
```

```python
chi2_stat
```

```python
p_value
```

```python
print("Observed  Expected  Difference") 
print("-" * 45)
for obs, exp in zip(observed, expected): 
    print(f"{obs:>8}  {exp:>8.0f}  {obs - exp:>+10.0f}") 
```

### Task 2: Chi-Square Test of Independence 

```python
table = np.array([[90, 60, 30],
                [110, 140, 170]]) 
```

```python
chi2, p, df, expected = st.chi2_contingency(table) 
```

```python
expected_df = pd.DataFrame(
    expected, index=["Churned", "Retained"], columns=["Free", "Basic", "Premium"] 
)  
```

```python
expected_df 
```

```python
chi2
```

```python
df
```

```python
p
```

There is evidence of an association between subscription tier and churn.

### Task 3: Compute Cramér's V

```python
def cramers_v(chi2, n, min_dim):
    return np.sqrt(chi2 / (n * min_dim))

n = table.sum()          
min_dim = min(table.shape) - 1  

v = cramers_v(chi2, n, min_dim)
```

```python
v
```

The relationship between subscription tier and churn is moderate.

It’s not just random — but also not extremely strong
